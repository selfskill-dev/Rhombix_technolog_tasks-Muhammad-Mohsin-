```python
# Import necessary libraries

import pandas as pd
import numpy as np

```

```python
# Lets we can create three datasets

# sales dataset
sales_data = pd.DataFrame({
    'sale_id': [1, 2, 3, 4, 5],
    'product_id': [101, 102, 103, 101, 104],
    'customer_id': [1001, 1002, 1003, 1001, 1004],
    'quantity': [2, 1, 3, 2, 5],
    'sale_price': [500, 300, 450, 500, 700],
    'cost_price': [350, 200, 300, 350, 500]
})
```

```python
product_data = pd.DataFrame({
    'product_id': [101, 102, 103, 104],
    'product_name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'category': ['Electronics', 'Accessories', 'Accessories', 'Electronics']
})

```

```python

customer_data = pd.DataFrame({
    'customer_id': [1001, 1002, 1003, 1004],
    'customer_name': ['Ali', 'Sara', 'Ahmed', 'Fatima'],
    'region': ['Lahore', 'Karachi', 'Islamabad', 'Lahore']
})
```

```python
merged_data = sales_data.merge(product_data, on='product_id', how='left')
merged_data = merged_data.merge(customer_data, on='customer_id', how='left')

print(merged_data)
```

```python
# Total Revenue per sale
merged_data['total_revenue'] = merged_data['quantity'] * merged_data['sale_price']

# Total Cost per sale
merged_data['total_cost'] = merged_data['quantity'] * merged_data['cost_price']

# Profit
merged_data['profit'] = merged_data['total_revenue'] - merged_data['total_cost']

# Profit Margin (%)
merged_data['profit_margin'] = (merged_data['profit'] / merged_data['total_revenue']) * 100
```

```python
print(merged_data)
```

```python
merged_data.to_csv('final_structured_data.csv', index=False)
```
